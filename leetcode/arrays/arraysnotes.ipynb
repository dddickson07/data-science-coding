# Arrays Data Structure

### General Notes

- Lists are a data structure used to store multiple elements
- Used in many algorithms
- Created using square brackets []
- List methods include append, sort, etc.
  - The different list methods are:
    - append() - adds an element at the end of the list
    - clear() - removes all the elements from the list
    - copy() - returns a copy of the list
    - extend() - Adds the elements of a list to the end of the current list
    - index() - Returns the index of the first element with the specified value
    - insert() - Adds an element at the specified position
    - pop() - removes the element at the specific position
    - reverse() - reverses the order of the list
    - sort() - sorts the list
- Create algorithms to perform actions not built into Python

```python
#Empty List
x = []

#list with initial values

y = [1,2,3,4,5]

# List with mixed types
z = [1, "hello",3.14, True]

## different methods
x = [9,12,7,4,11]
x.append(8) # adds an element
x.sort() #ascending sort
print(x)

#copy
y = x.copy()
print(y)

#count
nine_count = y.count(9)
print(nine_count)

#extend

y.extend(x)
print(y)

#index
index = y.index(9)
print(index)

y.insert(1, "pizza")
print(y)
y.pop(1)
print(y)
y.reverse()
print(y)

```

We create algorithms to perform actions not built into Python

```python
#example algorithm

my_array = [7,12,9,4,11,8]
minVal = my_array[0]

for i in my_array:
  if i < minVal:
    minVal = i

print('Lowest value:', minVal)
```
