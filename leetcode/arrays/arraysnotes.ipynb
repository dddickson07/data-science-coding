### General Notes

- Lists are a data structure used to store multiple elements
- Used in many algorithms
- Created using square brackets []
- List methods include append, sort, etc.

    - clear() - removes all the elements from the list

    - extend() - Adds the elements of a list to the end of the current list

###What are the benefits/limitations of arrays
- Constant time access to any value in the array (get and set are constant time)
- Arrays are not very dynamic (adding or removing an element in the middle means a lot need to be shifted, depends on n and i)
- Cannot expand or shrink; a new array needs to be allocated and data from old array needs to be copied into a new array... Expensive!
- Cost of growing and shrinking is O(m) over m operations

##**ArrayStack: Fast Stack Operations Using an Array**
Implemented using an array *a*, called the backing array. *a* is oftten times larger than itn needs to be.
- Integer *n* is used to keep track of the number of elements, it is at all times length(*a) >= n

###Operations
- initialize()
  - a --> new_array(1)
  - n --> 0
- get(i)
  - return a[i]
- set (i,x)
  - y <-- a[i]
  - a[i] <-- x
  - return y
- add(i,x)
  - if n = length(a) then resize()
    a[i+1, i+2, ...,n] <-- a[i,i+1,...,n-1] (shifts everything from i and beyond up one)
    a[i] <-- x
    n <-- n+1
  - runtime here is O(n-1)
- resize()
  - b <-- new_array(max(1,2*n)) (length because 2n or becomes 1 if it is 0)
  - b[0,1,...,n-1] <-- a[0,1,...,n-1]
  - a <--b
  - Takes O(n) time


*Lemma 2.1*: If an empty ArrayStack is created and any sequence of m >=1 is called to add (i,x) and remove (i) are performed, then the total time spent during all calls to resize() is O(m).
- essentialyl its because it saves a lot of time
- if you resize and to times two then you wont ened to resize it alot so the amortized one is going to be O(m) and not O(mn)

###Summary
- An ArrayStack implements the List interface and ignoring resize, it supports:
  - get(i) and set(i,x) in O(1) time per operation
  - add(i,x) and remove(i) in O(1 + n - i) time per operation

Additionally can implement push and pop!

##**ArrayQueue: An Array-Based Queue**

ArrayQueue data structure implements a FIFO queue; elements are removed from the queue in the same order they are added. ArrayStack is a bad choice because must choose one end of the list upon which to add elements and remove elements.


Problem would be eas if we had an infinite array; maintain a index j that keeps track of the next element to remove and n to count number of elements in the queue
  - a[j], a[j+1],..., a[j+n-1]
  - Add an element by placing it at a[j+n] and increment n, remove an element by placing it at a[j], increment j and decrement n
Modular arithmetic can simulate an infinite array since i mod length(a) always gives a value in the range.
  - store queue elements at a[j mod length (a)], a[(j+1) mod length (a)],...,[a(j+n-1)mod length(a)]

###Operations
- initialize()
  - a <-- new_array(1)
  - j <-- 0
  - n <-- 0

Check if a is full, resize to increase a, store x in a(j+n) mod length(a) and increment n.
- add(x)
  - if n+1 > length(a) then resize()
  - a[(j+n)mod length(a)] <-- x
  - n <-- n+1
  - return true
- remove()
  - x <-- a[j]
  - j <-- (j+1) mod length(a)
  - n <-- n - 1
  if length (a) >= 3 * n then resize()
  return x
- resize()
  - b <-- new_array(max(1,2*n))
    for k in 0,1,2...,n-1 do
      b[k] <-- a[(j+k) mod length(a)]
    a <-- b
    j <-- 0
