# Why Algorithms

Algorithms are used for everything in computer science. Things like opening a file, looking up a contact, logging into your social network, web search, phone emergency services. They are all completed by algorithms!

They make things **efficient**.


## Interfaces

Interace describes what a data structure does, implementation describes how the data structure does it.

Interfaces are also called abstract data type; define the set of operations supported by data structure, semantics/meaning of operations. Provides a list of supported operations and specificaations as well as what each type of argument and operation returns.

Data Structure implementation shows us **the internal representation.**

##Queue and Stack Interfaces
- add(x): add the value of x to a Queue
- remove(): remvoes the next (previously added) value from the Queue and returns y

###FIFO

First in first out queue. Removes items the same order they were added. Similar to a cash register or a grocery store. Most common type of queue.
- add(x)/enqueue(x)
- remove()/dequeue()

##Priority Queue
Removes smallest element from the Queue, breaking ties arbitrarly. Treats the patient with the most life-threatening condition. Often called delete_min() in other texts.
- add(x)
-remove()/delete_min()

##LIFO (last-in-first-out)
The most recently added element is the next one removed. A stack of plates.
- add(x)/push(x)
- remove()/pop()

## List Interface
- size(): return n, the length of the list
- get(i): return the value $x_i$
- set(i,x): set $x_i$ equal to x
- add (i,x): add x at position i
- remove(i) remove the value $x_i$, displacing all of the others

##The Unordered Set Interface:

- size():Returns the number, n, of elements in the set
- add(x): add the element x to the set if not already present
- remove(x): remove x from the set
- find(x): find x in the set if it exists

## The Sorted Set Interface:
- sorted so that x and y can be compared using compare(x,y)
- find(x) locates x in the sorted set
